```
import requests 
url = "https://listado.mercadolibre.com.co/colonias#D[A:colonias]"
response = requests.get (url)
print (response)
```

```
import bs4
parseo = bs4.BeautifulSoup (response.text, "html.parser")
ObtenerDatosSelect = lambda clase : parseo.select(clase)
titulos = ObtenerDatosSelect(".ui-search-item__title")
precios = ObtenerDatosSelect(".price-tag")
titulos
precios

funcionExtraerTexto = lambda elemento : elemento.text
funcionListaDatos = lambda funcion, data :  list (map (funcion,data))
nombresColonias = funcionListaDatos(funcionExtraerTexto, titulos)
listaPrecios = funcionListaDatos(funcionExtraerTexto, precios)
print(nombresColonias)
print (listaPrecios)
```

```
print(len(nombresColonias))
print(len(listaPrecios))
```

```
import pandas as pd
dictColonias = {}
dictColonias ["titulos"] = nombresColonias[:50]
dictColonias ["precios"] = listaPrecios[:50]
coloniasDataFrame = pd.DataFrame(dictColonias)
coloniasDataFrame.to_csv("Colonias.csv", index=False)
```

```
funcionAgregarColonia = lambda elemento : "Colonia : " + elemento
titulosColonias = list (map(funcionAgregarColonia,nombresColonias))
print (titulosColonias)
dictColonias = {}
dictColonias ["titulos"] = titulosColonias[:50]
dictColonias ["precios"] = listaPrecios[:50]
colonias = pd.DataFrame (dictColonias)
colonias.to_csv("ColoniasOrg.csv", index= False)  
```

```
dorsay = lambda elemento : "Dorsay" in elemento 
listaDorsay = list (filter(dorsay, titulosColonias))
dorsaySerie = pd.Series(listaDorsay)
dorsaySerie.to_csv("ColoniasDorsay.csv",  index_label="Dorsay")

nombresCortos = lambda elemento : len (elemento) <= 80
listaNombresFiltrados = list (filter (nombresCortos, listaDorsay))
print (listaNombresFiltrados)
```

```
from functools import reduce
magnitud = lambda elemento : len(elemento)
listaMagnitud = list(map(magnitud, titulosColonias))
print(listaMagnitud)
sumador = lambda acumulado = 0, elemento = 0: acumulado + elemento
promedio= reduce (sumador,listaMagnitud)/ len(titulosColonias)
promedio
```

```
import requests

url = "https://cnnespanol.cnn.com"
response = requests.get(url)
response?
print (dir(response))
```

```
import bs4
completo = bs4.BeautifulSoup(response.text,'html.parser')
noticias = completo.select('.news__title')
print (noticias)
```

```
titulos =[]
links = []
for noticia in noticias:
  titulos.append(noticia.text)
  links.append('https://cnnespanol.cnn.com/')
print (titulos)
print (links)
dictNoticias = {}
dictNoticias['Noticias']= titulos
dictNoticias['links']= links
```

```
import pandas as pd 
NoticiasCNNs= pd.DataFrame(dictNoticias)
print(NoticiasCNNs)
```

```
NoticiasCNNs.to_csv("noticia.csv", index= False)
```

```
import requests 
url = "https://listado.mercadolibre.com.co/televisor#D[A:televisor]"
response = requests.get (url)
print (response)
```

```
import bs4
parseo = bs4.BeautifulSoup (response.text, "html.parser")
ObtenerDatosSelect = lambda clase : parseo.select(clase)
titulos = ObtenerDatosSelect(".ui-search-item__title")
precios = ObtenerDatosSelect(".price-tag")
titulos
precios

funcionExtraerTexto = lambda elemento : elemento.text
funcionListaDatos = lambda funcion, data :  list (map (funcion,data))
nombresTelevisores = funcionListaDatos(funcionExtraerTexto, titulos)
listaPrecios = funcionListaDatos(funcionExtraerTexto, precios)
print(nombresTelevisores)
print (listaPrecios)
```

```
print(len(nombresTelevisores))
print(len(listaPrecios))
```

```
import pandas as pd
dictTv = {}
dictTv ["titulos"] = nombresTelevisores[:49]
dictTv ["precios"] = listaPrecios[:49]
tvDataFrame = pd.DataFrame(dictTv)
tvDataFrame.to_csv("Televisores.csv", index=False)
```

```
funcionAgregarNombre = lambda elemento : "Nombre : " + elemento
titulosTVs = list (map(funcionAgregarNombre,nombresTelevisores))
print (titulosTVs)
dictTv = {}
dictTv ["Titulos"] = titulosTVs[:49]
dictTv ["Precios"] = listaPrecios[:49]
televisores = pd.DataFrame(dictTv)
televisores.to_csv('TelevisoresOrg.csv', index= False) 
```

```
tedge = lambda elemento : "Tedge" in elemento 
listaTedge = list (filter(tedge, titulosTVs))
print(listaTedge)
tedgeSerie = pd.Series(listaTedge)
tedgeSerie.to_csv("TelevisoresTedge.csv",  index_label="Tedge")
```

```
from functools import reduce
sumador = lambda acumulado = 0, elemento = 0: acumulado + elemento
resultado = reduce (sumador,titulosTVs)
print (resultado)
```

```

nombresCortos = lambda elemento : len (elemento) <= 25
listaNombresFiltrados = list (filter (nombresCortos, nombresTelevisores))
print (listaNombresFiltrados)
```

```
from functools import reduce
magnitud = lambda elemento : len(elemento)
listaMagnitud = list(map(magnitud, titulosTVs))
print(listaMagnitud)
promedio= reduce (sumador,listaMagnitud)/ len(titulosTVs)
promedio
```
